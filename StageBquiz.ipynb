```python
import pandas as pd
import numpy as np

df = pd.read_csv('energydata_complete.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df1 = df.drop(columns=['date','lights'])
df1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df1 = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)
features_df1 = normalised_df1.drop(columns=['Appliances'])
target_Appliances = normalised_df1['Appliances']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(features_df1, target_Appliances, test_size=0.3, random_state=42)

```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
predicted_values = linear_model.predict(x_test)
predicted_values
```

```python
linear_model.coef_
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 3)
```

```python
#R_2
from sklearn.metrics import r2_score
r2_score1 = r2_score(y_test, predicted_values)
round(r2_score1, 2)
```

```python
#RSS
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 3)
```

```python
#RMSE
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
rmse
```

```python
#lasso Regression,  
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
pred_lasso = lasso_reg.predict(x_test)

Lasso_rmse = np.sqrt(mean_squared_error(y_test, pred_lasso))
lasso_r2_score = r2_score(y_test, pred_lasso)
lasso_mae = mean_absolute_error(y_test, pred_lasso)


print(round(Lasso_rmse,3))
print(lasso_r2_score)
print (lasso_mae)
print(lasso_reg.coef_)
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

```python
pred_ridge =ridge_reg.predict(x_test)

ridge_rmse = np.sqrt(mean_squared_error(y_test, pred_ridge))
ridge_r2_score = r2_score(y_test, pred_ridge)
ridge_mae = mean_absolute_error(y_test, pred_ridge)
print(ridge_rmse)
print(ridge_r2_score)
print(ridge_mae)
print(ridge_reg.coef_)
```

```python
def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_Weight')
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')

final_weights
```
